**Exercício 9.2**  
Escreva um programa que peça ao usuário para digitar várias linhas (até ele digitar "fim") e salve cada linha em um arquivo "notas.txt". Depois, leia o arquivo e exiba as linhas numeradas.


```python
# Programa para ler linhas e salvar em notas.txt
caminho = os.path.expanduser("/content/ARQUIVOS/syntaxpython/notas.txt")

# Cria diretórios se necessário
os.makedirs(os.path.dirname(caminho), exist_ok=True)

# Solicita as linhas ao usuário
print("Digite suas linhas (digite 'fim' para encerrar):")

with open(caminho, "w", encoding="utf-8") as arquivo:
    while True:
        linha = input()
        if linha.lower() == "fim":
            break
        arquivo.write(linha + "\n")  # adiciona quebra de linha

# Agora lê o arquivo e exibe com numeração
print("\nConteúdo do arquivo notas.txt:")
with open(caminho, "r", encoding="utf-8") as arquivo:
    linhas = arquivo.readlines()

for i, linha in enumerate(linhas, start=1):
    print(f"{i}: {linha.rstrip()}")  # rstrip remove a quebra de linha extra
```

```python
texto = "Python é legal"
```

```python
def reverte(txt):
    palavras = texto.split()
    palavras.reverse()
    return ' '.join(palavras)
```

```python
print(reverte(texto))
```
