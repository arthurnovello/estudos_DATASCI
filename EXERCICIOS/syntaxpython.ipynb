**Exercício 2.1**

```python
# Texto
texto = "Texticulo é um pequeno texto"

# Numerico
inteiro = 12
flutuante = 1.23
complexo = 42j

# Sequencias
lista = ['a', 'b', 'c']
tupla = ('alpha', 'beta', 'gamma')
intervalo = range(25)

# Mapa
dicionario = {"nome": "Aurelio", "capa": "Azul"}

# Conjunto
conjunto = {'m', 'n', 'o'}
conjuntoCongelado = frozenset({'i', 'j', 'k'})

# Booleano
booleano = True

# Binario
osBytes = b"Arthur"
arrayDosBytes = bytearray(25)
memoria = memoryview(bytes(5))
```

**Exercício 5.3**  
Crie uma função que receba uma string e retorne um dicionário com a contagem de cada caractere (use um loop e métodos de string, ou `count()` dentro de um set). Exemplo: "banana" -> {'b': 1, 'a': 3, 'n': 2}.

```python
def strToDict(string):
    dicionario = {}

    # varre a string e atualiza o dicionario
    for i in range(len(string)):
        dicionario.update({string[i]: i+1})

    return dicionario
```

```python
print(strToDict(texto))
```
